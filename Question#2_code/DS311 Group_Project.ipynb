```python
# Q2 I. What states (of those I am willing to move to) have the highest paying data-related salaries?
#States willing to move to: Texas, Washington, Alabama and will include California <- WORK_ST 

WORK_ST = ['Alabama', 'Washington', 'Texas', 'California']

# data-related jobs: Software Engineer, Senior software Engineer <- JOBT

JOBT = ['SOFTWARE ENGINEER','SENIOR SOFTWARE ENGINEER']

df.loc[df['JOB_TITLE'].isin(JOBT),['WORK_STATE',
                                   'JOB_TITLE_SUBGROUP','PAID_WAGE_PER_YEAR',
                                   'EMPLOYER_NAME']].sort_values(by="PAID_WAGE_PER_YEAR", 
                                                                            
                                                                 ascending=False).loc[df['WORK_STATE'].isin(WORK_ST)]
                                                                                                
#Can include .head() to adjust list      
```

```python
# Differences between job sub-categories?

#Data related jobs; subcategory: data scientist, business analyst, software engineer, data analyst

Data_related_jobs = ['data scientist','business analyst','software engineer', 'data analyst']
df.loc[df['JOB_TITLE_SUBGROUP'].isin(Data_related_jobs),['WORK_STATE','JOB_TITLE_SUBGROUP',
                                                         'PAID_WAGE_PER_YEAR',
                                                         'EMPLOYER_NAME']].sort_values(by="PAID_WAGE_PER_YEAR", 
                                                                            
                                                                 ascending=False).loc[df['WORK_STATE'].isin(WORK_ST)].head(21)
```
